```python
# 관광지별 사진 개수

tour_len = []

for nation in nation_list:
    for tour in tour_li:
        if os.path.exists(f'img/{nation}/{tour}'):
            ab = f'D:\\B-Project3\\B-Project3\\img\\{nation}\\{tour}'
            tour_len.append(len(os.listdir(ab)))
tour_len
```

```python
# 관광지별 사진 40% 개수

tour_four = []

for q in tour_len:
    tour_four.append(int(q * 0.4))

tour_four
```
