```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import os
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import MaxPooling2D, Conv2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
from tensorflow.image import central_crop, pad_to_bounding_box, resize
```

```python
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
  # Restrict TensorFlow to only allocate 1GB of memory on the first GPU
    try:
        tf.config.experimental.set_virtual_device_configuration(
            gpus[0],
            [tf.config.experimental.VirtualDeviceConfiguration(memory_limit=1024)])
        logical_gpus = tf.config.experimental.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
        # Virtual devices must be set before GPUs have been initialized
        print(e)
```

```python
config = tf.compat.v1.ConfigProto()
config.gpu_options.allow_growth = True
config.gpu_options.per_process_gpu_memory_fraction = 0.4 
session = tf.compat.v1.Session(config=config)
```

```python
from tensorflow.python.client import device_lib
print(device_lib.list_local_devices())
```

```python
df = pd.read_excel('data_file/여행지 라벨.xlsx')
df.head()
```

```python
tours = list(df[:20]['en관광지'].values)
tours
```

```python
countrys = df[:5]['en나라'].values
tours = df[:5]['en관광지'].values
```

```python
train_datagen = ImageDataGenerator(rescale=0./255)
validataion_datagen = ImageDataGenerator(rescale=0./255)
test_datagen = ImageDataGenerator(rescale=0./255)
```

```python
target_size = (256, 256)
batch_size = 20
train_generator = train_datagen.flow_from_directory('img/img_gen/train', target_size = target_size, batch_size = batch_size, classes=tours, shuffle=True)
validataion_generator = validataion_datagen.flow_from_directory('img/img_gen/validataion', target_size = target_size, batch_size = batch_size, classes=tours)
test_generator = test_datagen.flow_from_directory('img/img_gen/test', target_size = target_size, batch_size = batch_size, classes=tours)
```

```python
for x_data, y_data in train_generator:
    print(x_data.shape)  # (20, 150, 150, 3)
    print(type(x_data))  # <class 'numpy.ndarray'>
    print(y_data)
    break
```

```python
fig, axes = plt.subplot(10,2, figsize=(10,10))
for i in range(20):
    fig.imshow(x_data[i]/255)
```

```python
x = []
y = []
size = (256, 256)
countrys = os.listdir('img')
for country in countrys:
    names = os.listdir(f"img/{country}")
    for name in names:
        for i in range(100):
            path = f"img/{country}/{name}/{i}.jpg"
            img = load_img(path)
            img_array = img_to_array(img) / 255
            img_resize = resize(img_array, size)
            x.append(np.array(img_resize))
            y.append(name)
        print(f"{name} transform Successful")
x = np.array(x)
```

```python
plt.imshow(x[0])
```

```python
x_split = x[:500]
y_split = y[:500]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_split, y_split, test_size=0.2, random_state=111)
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=111)
```

```python
en = LabelEncoder()
y_train_en = en.fit_transform(y_train)
y_test_en = en.fit_transform(y_test)
```

```python
y_train_oh = to_categorical(y_train_en)
y_test_oh = to_categorical(y_test_en)
```

```python
y_test_oh.shape
```

```python
with tf.device('/GPU:0'):
    model = Sequential()
    model.add(Conv2D(32, (3,3), strides=1, activation='relu', padding='same', input_shape=(256, 256, 3)))
    model.add(MaxPooling2D(4))
    model.add(Dropout(0.2))

    model.add(Conv2D(64, (3,3), strides=1, activation='relu', padding='same'))
    model.add(MaxPooling2D(4))
    model.add(Dropout(0.2))

    model.add(Flatten())
    
    model.add(Dense(256, activation='relu'))
    model.add(Dense(20, activation='softmax'))
    model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
# with tf.device('/GPU:0'):
#     hist = model.fit(x_train, y_train_oh, batch_size=10, epochs=50, validation_data=(x_test, y_test_oh))
```

```python
with tf.device('/GPU:0'):
    hist = model.fit_generator(
        train_generator,
        steps_per_epoch=20,
        epochs=200,
        validation_data=validataion_generator,
        validation_steps=10,
)
```

```python
plt.figure(figsize=(10,10))
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
```

```python
model.evaluate(x_test, y_test_oh)
```

```python
pred = model.predict(x_test)
```
